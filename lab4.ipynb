# Варіант 3

\begin{cases}
8.2 x_1 + 7.9 x_2 = 123 \\
11.2 x_1 + 16.3 x_2 = 201 \\
-16.3 x_1 - 8.3 x_2 = 169 \\
\end{cases}

```python
import numpy as np
import matplotlib.pyplot as plt

def line1(x):
    return (123 - 8.2 * x) / 7.9

def line2(x):
    return (201 - 11.2 * x) / 16.3

def line3(x):
    return (169 + 16.3 * x) / -8.3

x_values = np.linspace(-20, 20, 400)

plt.figure(figsize=(15, 5))

plt.subplot(311)
plt.plot(x_values, line1(x_values), label='8.2x + 7.9y = 123', color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(312)
plt.plot(x_values, line2(x_values), label='11.2x + 16.3y = 201', color='green', linestyle=':')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(313)
plt.plot(x_values, line3(x_values), label='-16.3x - 8.3y = 169', color='blue', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.savefig("lab4.jpg")
plt.savefig("lab4.png")
plt.savefig("lab4.svg")
plt.show()
```

```python
import math

coefficients = np.array([[8.2, 7.9], [11.2, 16.3], [-16.3, -8.3]])
rhs = np.array([123, 201, 169])

x1, y1 = np.linalg.solve([coefficients[0], coefficients[1]], [rhs[0], rhs[1]])
x2, y2 = np.linalg.solve([coefficients[2], coefficients[1]], [rhs[2], rhs[1]])
x3, y3 = np.linalg.solve([coefficients[0], coefficients[2]], [rhs[0], rhs[2]])

x_vals = [x1, x2, x3, x1]
y_vals = [y1, y2, y3, y1]
print(x_vals, y_vals)
x_values = np.linspace(-50, 10, 400)
plt.ylim(5, 80)
plt.plot(x_values, line1(x_values), label='8.2x + 7.9y = 123', color='red', linestyle='--')
plt.plot(x_values, line2(x_values), label='11.2x + 16.3y = 201', color='green', linestyle=':')
plt.plot(x_values, line3(x_values), label='-16.3x - 8.3y = 169', color='blue', linestyle='-')


plt.plot(x1, y1, 'bo', label='Intersection 1', markersize=10)
plt.plot(x2, y2, 'ro', label='Intersection 2', markersize=10)
plt.plot(x3, y3, 'go', label='Intersection 3', markersize=10)


plt.fill_between(x_vals, y_vals, color='blue', alpha=0.3)
plt.legend()

plt.show()
```
